```python
import torch
import pandas as pd
from sklearn.model_selection import train_test_split
from torch import nn
from pathlib import Path
import numpy as np
from torchmetrics.regression import MeanAbsolutePercentageError
device = "mps" if torch.backends.mps.is_available() else "cpu"
```

```python
df = pd.read_csv('../excel_files/df_sparsed_train.csv')

df
```

```python
X = df.drop(columns=["373", "396"])

y = df["396"]
```

```python
X
```

```python
y
```

```python
X_tensor = torch.tensor(X.values, dtype=torch.float).to(device)
y_tensor = torch.tensor(y.values, dtype=torch.float).to(device)

X_tensor.shape, y_tensor.shape, len(X_tensor), len(y_tensor), X_tensor.device, y_tensor.device
```

```python
RANDOM_SEED = 42
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=RANDOM_SEED)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, len(X_train), X_train.device
```

```python
X_train.shape[1]
```

```python
class HousePricingModelV0(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_stack = nn.Sequential(
            nn.Linear(in_features=X_train.shape[1], out_features=512),
            nn.ReLU(),
            nn.ReLU(),
            nn.Dropout(p=0.3),
            nn.Linear(in_features=512, out_features=16),
            nn.ReLU(),
            nn.ReLU(),
            nn.Linear(in_features=16, out_features=1),
        )
    def forward(self, x):
        return self.layer_stack(x)
```

```python
torch.manual_seed(RANDOM_SEED)
model_0 = HousePricingModelV0()
model_0.to(device)
model_0
```

```python
y_pred_temp = model_0(X_train)

y_pred_temp
```

```python
loss_fn = nn.L1Loss()
optimizer = torch.optim.Adam(params=model_0.parameters(), lr=0.001)
mean_abs_percentage_error = MeanAbsolutePercentageError().to(device)
```

```python
torch.manual_seed(RANDOM_SEED)
epochs = 50000
for epoch in range(epochs):
    model_0.train()
    y_pred = model_0(X_train).squeeze()
    loss = loss_fn(y_pred, y_train)
    mape = mean_abs_percentage_error(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    

    model_0.eval()
    with torch.inference_mode():
        test_pred = model_0(X_test).squeeze()
        test_loss = loss_fn(test_pred, y_test)
        test_mape = mean_abs_percentage_error(test_pred, y_test)
    if (epoch + 1) % 1000 == 0:
        print(f"Epoch: {epoch+1}/{epochs} | {loss = :.2f} | {mape = :.2f} | {test_loss = :.2f} | {test_mape = :.2f}")
```
