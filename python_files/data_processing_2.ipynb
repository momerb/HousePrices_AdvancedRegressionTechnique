```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
```

```python
df1 = pd.read_csv('../excel_files/train.csv')
df2 = pd.read_csv('../excel_files/test.csv')

df = pd.concat([df1, df2], ignore_index=True)

df
```

```python
df.columns
```

```python
columns_to_fill = ['LotFrontage', 'GarageYrBlt', 'GarageArea', 'BsmtUnfSF', 'TotalBsmtSF'] 

median_values = df[columns_to_fill].median()

df[columns_to_fill] = df[columns_to_fill].fillna(median_values)

df
```

```python
missing_values = df[columns_to_fill].isnull().sum()

missing_values = missing_values[missing_values == 0]

missing_values
```

```python
columns_to_fill = ['MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtFullBath', 'BsmtHalfBath'] 

df[columns_to_fill] = df[columns_to_fill].fillna(0)

df
```

```python
missing_values = df[columns_to_fill].isnull().sum()

missing_values = missing_values[missing_values == 0]

missing_values
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].replace(2207, df['GarageYrBlt'].median())
df
```

```python
df['GarageCars'] = df['GarageCars'].fillna(2)

df['GarageCars'].isnull().sum()
```

```python
df = df.drop(columns=['YearBuilt'])

df.head()
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

len(numerical_columns), numerical_columns
```

```python
missing_values = df[numerical_columns].isnull().sum()

missing_values = missing_values[missing_values > 0]

missing_values
```

```python
df
```

```python
object_columns = df.select_dtypes(include=['object']).columns

len(object_columns), object_columns
```

```python
missing_values = df[object_columns].isnull().sum()

missing_values = missing_values[missing_values > 0]

missing_values.index.to_list()
```

```python
missing_values
```

```python
object_columns_with_missing_values = ['MSZoning',
 'Alley',
 'Utilities',
 'Exterior1st',
 'Exterior2nd',
 'MasVnrType',
 'BsmtQual',
 'BsmtCond',
 'BsmtExposure',
 'BsmtFinType1',
 'BsmtFinType2',
 'Electrical',
 'KitchenQual',
 'Functional',
 'FireplaceQu',
 'GarageType',
 'GarageFinish',
 'GarageQual',
 'GarageCond',
 'PoolQC',
 'Fence',
 'MiscFeature',
 'SaleType']

df[object_columns_with_missing_values] = df[object_columns_with_missing_values].fillna('NotPresent')

df
```

```python
missing_counts = df.isnull().sum()

columns_with_missing_values = missing_counts[missing_counts > 0].index.tolist()

columns_with_missing_values
```

```python
df['AgeCalculated'] = df['YrSold'] - df['YearRemodAdd']

df
```

```python
df.columns
```

```python
columns_to_scale = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 
                    'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'GarageArea',
                    'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 
                    'PoolArea', 'MiscVal', 'YearRemodAdd', 'YrSold', 'GarageYrBlt', 'AgeCalculated']

scaler = MinMaxScaler()

df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

df
```

```python
df.to_csv('../excel_files/data_check_1.csv', index=False)
```

```python
df
```

```python
categorical_columns = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
                        'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
                        'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
                        'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
                        'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                        'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
                        'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
                        'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
                        'SaleType', 'SaleCondition', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 
                        'HalfBath', 'TotRmsAbvGrd', 'Fireplaces', 'GarageCars', 'MoSold',
                        'MSSubClass', 'OverallQual', 'OverallCond', 'BedroomAbvGr', 'KitchenAbvGr']

one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', one_hot, categorical_columns)], remainder="passthrough")

transformed_df = transformer.fit_transform(df)

transformed_df
```

```python
dense_array_df = transformed_df.toarray()

dense_array_df
```

```python
df_sparsed = pd.DataFrame(dense_array_df)

df_sparsed
```

```python
df_sparsed_train = df_sparsed.iloc[:1460]
df_sparsed_test = df_sparsed.iloc[1460:]

df_sparsed_train[373]
```
